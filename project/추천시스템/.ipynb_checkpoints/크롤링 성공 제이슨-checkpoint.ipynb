```python
대분류_list = []
문의및안내_list = []
소분류_list = []
시군_list = []
읍면동_list = []
정보명_list = []
주소_list = []
중분류_list = []
지역_list = []
```

```python
for x in json_total['data']:
    #print (x['대분류'])
    대분류_list.append(x['대분류'])
    문의및안내_list.append(x['문의및안내'])
    소분류_list.append(x['소분류'])
    시군_list.append(x['시군'])
    읍면동_list.append(x['읍면동'])
    정보명_list.append(x['정보명'])
    주소_list.append(x['주소'])
    중분류_list.append(x['중분류'])
    지역_list.append(x['지역'])
    
len(대분류_list)
```

```python
import pandas as pd
```

```python
region_df = pd.DataFrame({'대분류':대분류_list,
                         '문의및안내':문의및안내_list,
                         '소분류':소분류_list,
                         '시군':시군_list,
                         '읍면동':읍면동_list,
                         '정보명':정보명_list,
                         '주소':주소_list,
                         '중분류':중분류_list,
                         '지역':지역_list})
```

```python
region_df 
```

```python
region_df.to_csv('region.csv', encoding = 'utf-8') 
```
