```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/Users/syshin/Desktop/Himedia/Himedia_project/문화체육관광부/2017 국내여행/0. 응답자 특성/1. 응답자 특성_개인.xlsx')
df
```

```python
df.columns
```

```python
use_col = [
    'PID', ## userID
    'sex', ## 1. 남성, 2.여성
    'age',
    'sido', ## 거주 시도 1.서울, 2.부산, 3.대구, 4.인천, 5.광주, 6. 대전, 7. 울산, 8. 경기, 9. 강원, 10.충북, 11.충남, 12.전북, 13.전남, 14.경북, 15.경남, 16. 제주
    'ara_size', ## 지역 규모 1. 대도시 2.중소도시, 3. 읍/면
    'occ1', ## 직업분류
    'fac', ##가구내 가구원 수
    'marry', ## 1. 미혼, 2. 배우자 있음, 3. 사별, 4. 이혼
    'inc1_1' ## 연봉
]
```

```python
df = df[use_col]
df
```

```python
df.dropna(axis=0, how='any', subset=use_col, inplace=True)
```

```python
df = df.rename(columns={
    'PID':'userID', ## userID
    'sido':'living_area', ## 거주 시도 1.서울, 2.부산, 3.대구, 4.인천, 5.광주, 6. 대전, 7. 울산, 8. 경기, 9. 강원, 10.충북, 11.충남, 12.전북, 13.전남, 14.경북, 15.경남, 16. 제주
    'ara_size':'living_area_size', ## 지역 규모 1. 대도시 2.중소도시, 3. 읍/면
    'occ1':'occupation', ## 직업분류
    'fac':'family_size', ##가구내 가구원 수
    'inc1_1':'income_per_year' ## 연봉
})
```

```python
df
```

```python
living_area_dic={
    1: '서울특별시', 
    2: '부산광역시', 
    3: '대구광역시', 
    4: '인천광역시', 
    5: '광주광역시', 
    6: '대전광역시', 
    7: '울산광역시', 
    8: '경기도', 
    9: '강원도', 
    10: '충청북도', 
    11: '충청남도', 
    12: '전라북도', 
    13: '전라남도', 
    14: '경상북도', 
    15: '경상남도', 
    16: '제주특별자치도'
}
```

```python
living_area_size_dic={
    1: '대도시', 
    2: '중소도시', 
    3: '읍/면'
}
```

```python
occupation_dic={
    1:'관리자',
    2:'전문가 및 관련 종사자',
    3:'사무 종사자',
    4:'서비스 종사자',
    5:'판매 종사자',
    6:'농림어업 숙련 종사자',
    7:'기능원 및 관련 기능 종사자',
    8:'강치/기계 조작 및 조립 종사자',
    9:'단순노무자',
    10:'국인',
    11:'가정주부',
    12:'중학생',
    13:'고등학생',
    14:'대학생',
    15:'대학원생',
    16:'은퇴/무직',
    17:'무응답'
}
```

```python
marry_dic={
    1: '미혼', 
    2: '배우자 있음', 
    3: '사별', 
    4: '이혼'
}
```

```python
for index in df.index:
    ## 성별 string으로 입력
    if df.loc[index,'sex']==1: df.loc[index,'sex']='남성'
    else: df.loc[index,'sex']='여성'
    
    df.loc[index,'living_area'] = living_area_dic[df.loc[index,'living_area']]
    df.loc[index,'living_area_size'] = living_area_size_dic[df.loc[index,'living_area_size']]
    df.loc[index,'occupation'] = occupation_dic[df.loc[index,'occupation']]
    df.loc[index,'marry'] = marry_dic[df.loc[index,'marry']]
df    
```

```python
df.to_csv('user_trip.csv')
```
