```python
list1 = ['지역호수/저수지', '지역사찰', '일반관광지', '대형호수/저수지', '숲']
list2 = ['폭포/계곡', '관광농원/허브마을', '해수욕장', '일반유원지/일반놀이공원', '테마공원/대형놀이공원', '아쿠아리움/대형수족관', '카지노', '파라다이스카지노', '세븐럭카지노']
list3 = ['토속/특산물/기념품매장', '먹거리/패션거리']
list4 = ['지역축제', '서원/향교/서당', '휴양림/수목원', '유명사찰', '유명관광지', '식물원', '동물원']
```

```python
for index in df.index:
    if df.loc[index,'mcate_nm'] in list1:
        df.loc[index,'target']='힐링, 호캉스'
    elif df.loc[index,'mcate_nm'] in list2:
        df.loc[index,'target']='액티비티'
    elif df.loc[index,'mcate_nm'] in list3:
        df.loc[index,'target']='식도락, 맛집투어'
    elif df.loc[index,'mcate_nm'] in list4:
        df.loc[index,'target']='축제, 관광'
```
