```python
plt.figure(figsize=(15,6))
sns.countplot('Sex', data=raw_data)
plt.show()
```

```python
# 성별 질환인원 구해보고, 그래프로 그려보세요!
sex_alive = raw_data.groupby(by='Sex')['HeartDisease'].sum() #1을 다 더하면, 생존인원이 될 것임.
print(sex_alive)
list_sa = list(sex_alive)
list_sa.reverse()
list_sa ##생존인원
```

```python
print(rd_sex) 
rd_sex_list = list(rd_sex)
rd_sex_list
```

```python
for i in range(2):
    print(rd_sex.index[i])
    print('patient인원: ',list_sa[i], ' <-> total인원: ',  rd_sex_list[i])
    print('----------------------------------------')
```

```python
sex_df = pd.DataFrame({'patient' : list_sa, 'total' : rd_sex_list}, index=['male', 'female'])
sex_df
```

```python
patient_result = sex_df['total'] - sex_df['patient']
patient_result
```

```python
sex_df.reset_index(inplace=True)
```

```python
f,ax=plt.subplots(1,2,figsize=(12,6))
sns.barplot('index', 'total', data=sex_df, ax=ax[0])
sns.barplot('index', 'patient', data=sex_df, ax=ax[1])
```

```python
plt.figure(figsize=(7, 6))
plt.plot(sex_df['index'], sex_df['total'], color='g')
plt.plot(sex_df['index'], sex_df['patient'], color='r')
plt.show()
```
