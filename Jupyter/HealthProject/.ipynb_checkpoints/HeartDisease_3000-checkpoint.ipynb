```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
# warnings.filterwarnings(action='default')
%matplotlib inline
```

```python
raw_data = pd.read_excel("heart_2020_3000.xlsx")
raw_data
```

```python
raw_data.info()
```

```python
raw_data.isnull().sum()
```

```python
raw_data['BMI'] = raw_data['BMI'].astype(float)
raw_data.info()
```

```python
x = raw_data.loc[:, 'BMI':]
x
```

```python
y = raw_data.loc[:,'HeartDisease']
y
```

```python
import random
random.seed(42) #시드 == 씨앗, 씨앗값
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x,  #문제
                        y, #정답
                        test_size=0.2, #검증비율 20%, 훈련비율 80%
                        random_state=100 #seed
                )
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
y_train.count()
```

```python
y_test.count()
```

```python
tree_clf = DecisionTreeClassifier(max_depth=8, random_state=22)
```

```python
tree_clf.fit(X_train, y_train)
```

```python
tree_clf.score(X_train, y_train)
```

```python
com_predict = tree_clf.predict(X_test)
```

```python
accuracy_score(y_test, com_predict)
```

```python
X_train.columns
```

```python
y_train.value_counts()
```

```python
feature_importance = pd.DataFrame({
    'feature': x.columns,
    'importance': tree_clf.feature_importances_
})
feature_importance
```

```python
tree_clf.feature_importances_
```

```python
result = tree_clf.predict_proba([[20.8,0,0,1,30,30,1,0,30,0,0,0,1,6,0,0,0]])
result
```

```python
def func(r):
    if r[0][0] > r[0][1]:
            print('심장병이 있을 가능성이 더 높다.')
            print('가능성 지수는 ', int(r[0][0] * 100), '%')
    else:
            print('심장병이 없을 가능성이 더 높다.')
            print('가능성 지수는 ', r[0][1] * 100, '%')
```

```python
func(result)
```

```python
## 80퍼센트의 train, 20퍼의 test로 나누고 (max_depth=8, random_state=22)조건을 주었을때:
## feature_importances수치를 살펴보았을때, BMI수치가 가장 중요했고, 나이가 두번째로 중요했다. 육체적 건강, 일반적건강, 수면이 그 뒤를 이었고, 당뇨병 여부, 성별, 인종, 정신건강이 작은 영향력으로 그 뒤를 이었다.
## 위의 결과를 따르면, 의외로 담배나 음주는 심장병에는 큰 영향을 주지 않는 것으로 확인됐다
## 심장병이 있는 환자의 수치를 test해봤더니 100% 심장병을 예측했다
## 
```
