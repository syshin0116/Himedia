```python
from urllib import request
from bs4 import BeautifulSoup
con =request.urlopen('https://novel.naver.com/webnovel/ranking')
con 
```

```python

con.status #status code
```

```python
doc=BeautifulSoup(con,'html.parser')
doc
```

```python
def crawl_img:
    #이미지
    free_img=doc.select( '#content > div > div.ranking_lst > div.ranking_wrap_left > ul > li> a > div.pic > img ' )
    img_list=[]
    for i in range(0, len(free_img), 1):
        a= free_img[i].get('src')
        img_list.append(a)
    img_list

# len(img_list)
```

```python

#제목
free_title=doc.select( 'div.ranking_lst > div.ranking_wrap_left span.tit' )


free_title_list=[]
for i in range(0, len(free_title), 1):
    
    a= free_title[i].text
    aa =a.replace(',',' ')
    free_title_list.append(aa)

free_title_list
len(free_title_list)
```

```python
novel_file =open('webnovel_free.csv','w')
novel_file .write("title,img"+"\n")

for i in range(0,len (img_list),1):
    data=img_list[i]+","+free_title_list[i]+"\n"
    novel_file.write(data)
    
novel_file.close()  
```

```python
#별점
free_star=doc.select( ' div.ranking_wrap_left span.score_area' )

free_star_list=[]
for i in range(0, len(free_star), 1):
    a = free_star[i].text.replace('별점','')
    print(a)
    free_star_list.append(a)

len(free_star_list)
```

```python
a_tag_list = doc.select('div.ranking_wrap_left span.author')
# a_tag_list = doc.select('span.author')
a_tag_list
```

```python
for one in a_tag_list:
    print(one.text)
```

```python
author_list = []
for one in a_tag_list:
    author_list.append(one.text)
author_list
```

```python
free_author=doc.select('div.ranking_wrap_left   span.author')
free_author_list=[]
free_author
for i in range(0, len(free_author), 1):
    a= free_author[i].text
    print(a)
    free_author_list.append(a)

len(free_author_list)
```

```python
novel_file1 =open('naver_webnovel_free1.csv','w')
novel_file1 .write("img,title,author,star"+"\n")

for i in range(0,len (img_list),1):
    data=img_list[i]+","+free_title_list[i]+","+author_list[i]+","+free_star_list[i]+"\n"
    novel_file1.write(data)
    
novel_file1.close()  
```
