```python
from urllib import request
```

```python
from bs4 import BeautifulSoup
```

```python
con = request.urlopen('https://novel.naver.com/best/ranking?genre=101&periodType=DAILY')
con 
```

```python
con.status #status code
```

```python
doc = BeautifulSoup(con, 'html.parser')
```

```python
img_crawl = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div.pic > img')
img_crawl
```

```python
img_list =[]
```

```python
for crawl in img_crawl:
    img_list.append(crawl.get("src"))  # 이미지 태그
img_list
```

```python
len(img_list)
```

```python
title_list = []
```

```python
title_crawl = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p > span.tit')
title_crawl
```

```python
for title in title_crawl:
    title_list.append(title.text.replace(',',''))

title_list
```

```python
len(title_list)
```

```python
author_crawl = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p.sub_info > span.author')
author_crawl
```

```python
author_list = []
```

```python
for author in author_crawl:
    author_list.append(author.text.replace(',',''))
author_list
```

```python
len(author_list)
```

```python
rating_crawl = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p.rating > span.score_area')
rating_crawl
```

```python
rating_lists=[]
rating_list=[]
```

```python
for ratings in rating_crawl:
    rating_lists.append(ratings.text)
rating_lists
```

```python
for s in rating_lists:
    rating_list.append(s[2:])
rating_list
```

```python
len(rating_list)
```

```python
webtoon_file = open('webtoon.csv', 'a', encoding='utf8')
```

```python
for i in range(len(rating_list)):
    webtoon_file.write(title_list[i] + ',' +
                         author_list[i] + ',' +
                         img_list[i]+','+rating_list[i]+'\n')
```

```python
webtoon_file.close()
```
