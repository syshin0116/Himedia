```python
import warnings
warnings.filterwarnings(action='ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
# 검색할 키워드 입력
query = input("검색할 키워드를 입력하세요>>")
```

```python
# 크롬 드라이버로 원하는 url접속
# 1. 크롬 브라우저 실행
driver = webdriver.Chrome('C:/Users/hi/Himedia_jupyter/chromedriver.exe')

# 2. 특정 url: http://www.naver.com 로 연결
url = 'http://www.naver.com'
driver.get(url)
```

```python
# 연결하여 페이지를 받아오는 동안 기다리기 설정
time.sleep(0.5)
```

```python
# 3. 네이버 검색 input태그의 위치를 찾아서 
# #query
shopping_tab = driver.find_element_by_css_selector('a.nav.shop')
shopping_tab.click()
time.sleep(0.5)
search_box = driver.find_element_by_css_selector('#autocompleteWrapper > input.co_srh_input._input.N\=a\:SNB\.search')
search_box.send_keys(query)
# 4. 3번에서 찾은 input 태그에 검색어를 넣는다.
# 5. 4번에서 입력한 값 서버로 전송
search_box.send_keys(Keys.ENTER)
# 기다리기 설정
time.sleep(0.5)
price_tab = driver.find_element_by_css_selector('#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.style_content_wrap__1PzEo > div.style_content__2T20F > div.seller_filter_area > ul > li:nth-child(2) > a')
price_tab.click()
time.sleep(0.5)
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
```

```python
## 뉴스 제목 가지고 와보자 # #sp_nws1 > div.news_wrap.api_ani_send > div > a
item_brand = driver.find_elements_by_css_selector("a.basicList_brand_message__2-WlN")
item_title = driver.find_elements_by_css_selector('#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.style_content_wrap__1PzEo > div.style_content__2T20F > ul > div > div > li > div.basicList_inner__eY_mq > div.basicList_info_area__17Xyo > div.basicList_title__3P9Q7 > a')
item_price = driver.find_elements_by_css_selector('#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.style_content_wrap__1PzEo > div.style_content__2T20F > ul > div > div > li > div.basicList_inner__eY_mq > div.basicList_info_area__17Xyo > div.basicList_price_area__1UXXR > strong > span > span.price_num__2WUXn')
item_img = driver.find_elements_by_css_selector('#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.style_content_wrap__1PzEo > div.style_content__2T20F > ul > div > div > li > div > div.basicList_img_area__a3NRA > div > a  img')
print(len(item_price))
item_price
```

```python
title_list = []
brand_list = []
price_list = []
url_list = []
img_list = []
for brand in item_brand:
    brand_list.append(brand.text)
for title in item_title:
    title_list.append(title.text)
    url_list.append(title.get_attribute('href'))
for price in item_price:
    try:
        price = price.text.replace(',','')
        price = price.replace('원', '')
        price_list.append(price)
    except:
        price_list.append('판매종료')
for img in item_img:
    img_list.append(img.get_attribute('src'))
num = len(title_list)-(len(brand_list))
for i in range(num):
    brand_list.append(None)
print(len(title_list))
print(len(brand_list))
print(len(price_list))
print(len(url_list))
print(len(img_list))
price_list.insert(45, '0')
price_list
```

```python
import pandas as pd
import numpy as np
```

```python
naver_shop_df = pd.DataFrame({'title':title_list,
                              'brand':brand_list,
                              'price':price_list,
                                'url':url_list,
                                'img':img_list
                               })
naver_shop_df
```

```python
# 이미지 주소가지고img를 컴퓨터에 다운로드 해보자!
# 저장위치 결정, 현재 (폴더)경로 설정
path_folder = './img/' ## path_folder = './img'
import os
if not os.path.isdir(path_folder):
    os.mkdir(path_folder)

#이미지 다운로드
from urllib.request import urlretrieve
i = 0

for img in img_list:
    i+=1
    urlretrieve(img, path_folder + str(i)+ '.jpg')        #link에서 이미지 다운로드, './imgs/'에 파일명은 index와 확장자명으로
```
