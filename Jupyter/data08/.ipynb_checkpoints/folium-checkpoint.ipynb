```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../csv-data/folium-data.csv')
df.head()
```

```python
del df['no']
df.head()
```

```python
df.set_index('자치구', inplace=True)
```

```python
df.head()
```

```python
# 중기업, 소기업
map1 = folium.Map(location=[37.57, 126.98], 
                zoom_start=12
                )
map1.choropleth(geo_data=geo_str,
                      data= df['대기업'],
                      columns=[df.index, df['대기업']],
                      fill_color='PuRd',
                      key_on='feature.id'
              )


for n in df.index:
    folium.CircleMarker([df['lat'][n], df['lng'][n]], 
                        radius = df['대기업'][n]*0.1, 
                        color='#3186cc', fill_color='#3186cc', fill=True).add_to(map1)
map1
```

```python
##중기업, 소기업
map2 = folium.Map(location=[37.5310494040562, 126.99742645348145],
                zoom_start=11)
#마커를 찍어보자
folium.Marker([37.571289962964414, 126.98727395584191], popup='여기는 하이미디어', icon = folium.Icon(color='red', icon='info-sign').add_to(map))
map2.choropleth(geo_data=geo_str,
               data=df['중기업'],
               columns=[df.index, df['중기업']],
               fill_color='PuRd',
                key_on='feature.id'
              )
#서클을 그려보자
for n in df.index:
    folium.CircleMarker([df['lat'][n], df['lng'][n]], 
                        radius = df['중기업'][n]*0.1, 
                        color='#3186cc', fill_color='#3186cc', fill=True).add_to(map2)
map2
```

```python
##중기업, 소기업
map3 = folium.Map(location=[37.5310494040562, 126.99742645348145],
                zoom_start=10)
#마커를 찍어보자
folium.Marker([37.571289962964414, 126.98727395584191], popup='여기는 하이미디어', icon = folium.Icon(color='red', icon='info-sign').add_to(map))
map3.choropleth(geo_data=geo_str,
               data=df['소기업'],
               columns=[df.index, df['소기업']],
               fill_color='PuRd',
                key_on='feature.id'
              )
#서클을 그려보자

map3
```
