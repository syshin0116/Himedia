```python
total_row = []
for no in range(len(data_values)):
    clearVal = data_values[no]['clearVal']
    districtName = data_values[no]['districtName']
    dataDate = data_values[no]['dataDate']
    issueVal = data_values[no]['issueVal']
    issueTime = data_values[no]['issueTime']
    moveName = data_values[no]['moveName']
    clearTime = data_values[0]['clearTime']
    issueGbn = data_values[no]['issueGbn']
    itemCode = data_values[no]['itemCode']
    one_row = []
    one_row.append(clearVal)
    one_row.append(districtName)
    one_row.append(dataDate)
    one_row.append(issueVal)
    one_row.append(issueTime)
    one_row.append(moveName)
    one_row.append(clearTime)
    one_row.append(issueGbn)
    one_row.append(itemCode)
    print(one_row)
    total_row.append(one_row)
    print(' >> ', len(total_row))
```
