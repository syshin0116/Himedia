```python
## 컬럼간의 상관정도를 수치로 표현하여 컬럼간의 관계성을 고려해볼 수 있음.
## 상관도: 상관이 있는 정도
## 상관계수: 상관 정도가 수치로 표현된 것(corr())
```

```python
## 상관<------->연관
## 상관: 관계가 있긴 있음을 나타내는 정도
## 연관: 관계가 있고, 하나가 다른 하나에 영향을 미치는 정도
```

```python
# 상관도를 구할 떄 사용하는 방법
# 1) corr()함수를 이용
# 2) 그래프를 이용 = heatmape(), pairplot()
```

```python
df[['wine_servings','beer_servings']].corr()
```

```python
#상관도를 구하고 나서 수치보는 법
# 0.7~1: 높은 상관관계
# 0.5~0.7: 중간 상관관계
# 0.3~0.5: 낮은 상관관계
# 0~0.3: 아주 낮은 상관관계
# 상관계수가 + 이면, 양의 상관관계
# 상관계수가 - 이면, 음의 상관관계
```
