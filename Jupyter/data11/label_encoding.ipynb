```python
##라벨 인코딩 할 컬럼이 2-3개 인 경우
```

```python
words = ['apple', 'summer', 'summer', 'spring', 'spring']
hobby = ['book', 'run','talk','coffee','song']
```

```python
df = pd.DataFrame(
    {'words':words, 'hobby':hobby})
```

```python
encoder2 = LabelEncoder()
```

```python
col = df.columns
col
```

```python
for c in col:
    df[c] = encoder2.fit_transform(df[c])
```

```python
df
```

```python
# 나이(100, 200, 300): 연속적인 숫자 => 라벨인코딩의 대상이 되지 않음
# 나이대(어린이, 성년, 노년): 연속적인 숫자가 아님=>라벨인코딩의 대상이 됨. category
```

```python
### 원핫 인코딩 ###
### 라벨인코딩: 연속형이 아닌 데이터를 원하는 형태의 숫자로 바꾸어주는 것!
###            1) 연속형이 아닌 데이터를 숫자로 변환: LabelEncoder 사용
###               fit() -> transform(), fit_transform()
###            2) 숫자로만 변환하는 경우, 컴퓨터는 연속형 숫자로 오해하여 연산을 수행하여
###               의도치않은 결과가 나올 수 있음. 해당됨을 나타나게 하기 위해
###               라벨의 객수만큼 컬럼을 만들어주고 해당되는 라벨에 표시(1)로 해주는 방법을 사용
###               원핫인코딩
```

```python
df_drink = pd.read_csv('../csv-data/drinks.csv')
df_drink
```

```python
df_drink.info()
```

```python
# 1. beer_servings를 3개의 group으로 만드세요
df_drink['beer_servings'].value_counts().index.max()
```

```python
df_drink['continent'].value_counts()
```

```python
bcat_list = []
for i in df_drink['beer_servings']:
    if i<100:
        bcat_list.append('100>')
    elif i>=100 and i<200:
        bcat_list.append('100-200')
    else:
        bcat_list.append('200<')
len(bcat_list)
```

```python
# 2. 1번의 결과를 beer_cat변수로 추가
df_drink['beer_cat'] = bcat_list
df_drink.head()
```

```python
df_drink['beer_servings'][df_drink['beer_servings']>200]
```

```python
encoder3 = LabelEncoder()
```

```python
cat_series = df_drink['beer_cat']
```

```python
cat = ['100>', '100-200', '200<']
```

```python
encoder3.fit(cat)
```

```python
encoder3.transform(cat_series)
```

```python
encoder4 = LabelEncoder()
```

```python
#3. LabelEncoding fit_transform
encoder4.fit_transform(cat_series)
```

```python
# 4.핫인코딩 df전체
pd.get_dummies(df_drink)
```

```python
# 5. 원핫인코딩 특정 컬럼(1개 또는 그 이상)을 선택해서!
pd.get_dummies(df_drink['beer_cat'])
```

```python
df_drink['beer_cut']= pd.cut(
                            df_drink['beer_servings'],
                            bins=[0,100,200,400],
                            include_lowest=True,
                            labels=['low','mid','high']
                          )
df_drink[df_drink['beer_cut'] == 'low']
```

```python
df_drink['beer_cut2']= pd.qcut(
                            df_drink['beer_servings'],
                            3,
                            labels=['low','mid','high']
                          )
df_drink[df_drink['beer_cut2'] == 'low'].count
```
